```python
import pandas as pd
import numpy as np
data ={
    'ID':['1','2','3','4','5'],#duplicate ID + string type
    'Name':['Amit','Riya','Kiran', None,None], #missing value
    'Marks':[80,np.nan,75,75,np.nan], #missing values
    'Grade':['A','B','B',None,'B'] #missing + categorical
}

df=pd.DataFrame(data)
print(df)
print(df.isnull())
print(df.isnull().sum())
print(df.notnull())

print(df.dropna()) #drop rows with missing values
print(df.dropna(axis=1)) #drop columns with missing values
print(df.dropna(how='all')) #drop rows where all values are missing values
print(df.dropna(thresh=2)) #drop rows with at least 2 missing values
print(df.dropna(subset=["Name"]))
# df=df.dropna(subset=["Name"]) #creating copy
# df.dropna(subset=["Name"],inplace=True) #changing original index
print(df)
print(df['Marks'].fillna(df['Marks'].mean()))

print(df['Marks'].fillna(0))


print(df['Marks'].fillna(df['Marks'].median()))

print(df['Marks'].fillna(df['Marks'].mode()[0]))


print(df['Marks'].fillna(5))


print(df.duplicated())
print(df.drop_duplicates())

print(df['ID'].astype(int))

df['Marks']= pd.to_numeric(df['Marks'],errors='coerce')
print(df["Marks"])
```
